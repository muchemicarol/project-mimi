```python
def excel_to_dictionary(filename, sheetname):
    abbreviation_dict = dict()
    
    abbreviation = xlrd.open_workbook(os.path.abspath(filename))
    abbreviation_sheet = abbreviation.sheet_by_name(sheetname)
    
    for i in range(1, abbreviation_sheet.nrows):
        abb = abbreviation_sheet.cell(i, 0).value
        abb_definition = abbreviation_sheet.cell(i, 1).value
        abbreviation_dict[abb] = abb_definition
        
    return abbreviation_dict
```

```python
securities_dict = excel_to_dictionary(filename="PROJECT MIMI/DataDictionary.xlsx", sheetname="Securities")
ratings_dict = excel_to_dictionary(filename="PROJECT MIMI/DataDictionary.xlsx", sheetname="Ratings")
fundamentals_dict = excel_to_dictionary(filename="PROJECT MIMI/DataDictionary.xlsx", sheetname="Fundamentals")
```
